```python
b = a.groupby(['Order ID', 'Product Name'])['Quantity'].sum().unstack().fillna(0).reset_index().set_index('Order ID')
```

```python
b.head()
```

```python
b.describe()
```

```python
def encode(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

c = b.applymap(encode)
c.describe()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

itemset = apriori(c, min_support=0.002, use_colnames=True, verbose=1, max_len=2)
itemset
```
